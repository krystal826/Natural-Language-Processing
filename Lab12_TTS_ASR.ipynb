## 1.0 Text-to-Speech (using Google Speech API)

There are several APIs available to convert text to speech in Python. One of such APIs is the Google Text to Speech API commonly known as the gTTS API. gTTS is a very easy to use tool which converts the text entered, into audio which can be saved as a mp3 file.

The gTTS API supports several languages including English, Hindi, Tamil, French, German and many more. The speech can be delivered in any one of the two available audio speeds, fast or slow. However, as of the latest update, it is not possible to change the voice of the generated audio.


```python
# If tts module is yet available, type !pip install gTTS

import os
!pip install gtts
from gtts import gTTS

# The text that you want to convert to audio
mytext = 'Welcome to Natural Language Processing course!'
  
# Language in which you want to convert
language = 'en'
  
# Passing the text and language to the engine, 
# here we have marked slow=False. Which tells 
# the module that the converted audio should 
# have a high speed
myobj = gTTS(text=mytext, lang=language, slow=False)
  
# Saving the converted audio in a mp3 file named welcome 
myobj.save("welcome.mp3")
  
```

```python
engine = pyttsx3.init(); #default speech engine
engine.say("Welcome to Natural Language Processing course")
engine.runAndWait()
```

### Voices are different on different PCs. In some PCs, voice[0] is a male one, and voice[1] is a female one, but on another PC, there can be more voices, they can be both male/female etc. 

```python
voices = engine.getProperty('voices')

engine.setProperty('voice', voices[0].id) #change index to change voices
engine.say('Hello, good day!')

engine.runAndWait()
```

```python
engine = pyttsx3.init('sapi5') #use speeche engine sapi5 (for Windows) or nsss (for Mac)
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[1].id)
def speak(audio):
    engine.say(audio)
    engine.runAndWait()
speak('Hello World')
```

```python
recognizer = sr.Recognizer()

# Energy threshold = the loudness of the audio files
# The values below the threshold are considered silent
# The values above are considered speech 
recognizer.energy_threshold = 300 
```

### Audio Preprocessing
### Parameters: duration and offset, 
### duration = 7.0 means take only the forst 7 seconds of the whole audio file 
### offset = 1.0 means don't take the first second of the audio files

## Speak to the microphone and Google SR will convert into text 

```python
# Listen to what we said and display its corresponding text
r = sr.Recognizer()
with sr.Microphone() as source:
    print('Speak anything')
    audio = r.listen(source)
    print("Time over!")
    try:
        text = r.recognize_google(audio)
        print("You said : {}".format(text))
    except sr.UnknownValueError: 
        print("Google Speech Recognition could not understand audio") 
  
    except sr.RequestError as e: 
        print("Could not request results from Google Speech Recognition service; {0}".format(e))
```

```python
#listen to the wav file audio (en-0524.wav) and convert to the corresponding text
result = recognizer.recognize_google(audio_data=audio_file)
result
```

```python
#listen to the wav file audio (en-0524.wav), cut off the first second and 
#convert to the corresponding text

audio_file = sr.AudioFile("en-0524.wav")
with audio_file as source:
  audio_file = recognizer.record(source, offset = 2.0)
  result = recognizer.recognize_google(audio_data=audio_file)
result
```

```python
#listen to the wav file audio (en-0524.wav) and 
#convert to the list of possible corresponding text

audio_file = sr.AudioFile("en-0524.wav")
with audio_file as source:
  audio_file = recognizer.record(source)
  result = recognizer.recognize_google(audio_data=audio_file, show_all=True)
result
```

## Multi speakers

Audio files needed: speaker1.wav, speaker2.wav

```python
import IPython

# Playing the converted file
IPython.display.Audio("speaker1.wav")
```

```python
IPython.display.Audio("speaker2.wav")
```

```python
# Detect multi speakers
recognizer = sr.Recognizer()

# Multiple speakers on different files
speakers = [sr.AudioFile("speaker1.wav"), sr.AudioFile("speaker2.wav")]
# Transcribe each speaker individually
for i, speaker in enumerate(speakers):
    with speaker as source:
        speaker_audio = recognizer.record(source)
        print(f"Text from speaker {i+1}:")
        print(recognizer.recognize_google(speaker_audio,language="en"))
```

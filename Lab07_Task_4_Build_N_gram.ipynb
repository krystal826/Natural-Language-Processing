Lab 07 Task 4: Build a basic language model using trigrams of the Reuters corpus. 

Reuters corpus is a collection of 10,788 news documents totaling 1.3 million words. 

```python
# code courtesy of https://nlpforhackers.io/language-models/

from nltk.corpus import reuters    #using the trigrams of the Reuters corpus
from nltk import bigrams, trigrams
from collections import Counter, defaultdict
```

```python
# Create a placeholder for model
model = defaultdict(lambda: defaultdict(lambda: 0))
```

```python
# Count frequency of co-occurance  [w1, w2, w3]
for sentence in reuters.sents():
    for w1, w2, w3 in trigrams(sentence, pad_right=True, pad_left=True):
        model[(w1, w2)][w3] += 1
```

```python
# Let's transform the counts to probabilities
for w1_w2 in model:
    total_count = float(sum(model[w1_w2].values()))
    for w3 in model[w1_w2]:
        model[w1_w2][w3] /= total_count
```

1) Firstly, the text is split into trigrams with the help of NLTK and then calculate the frequency in which each combination of the trigrams occurs in the dataset.

2) Then use it to calculate probabilities of a word, given the previous two words. That's a Language Model!

3) Let's make simple predictions with this language model. Start with two simple words – “today the”.

Here is a script to play around with generating a random piece of text using the n-gram model.

The model generate a sequence of text after "today the"

```python
# code courtesy of https://nlpforhackers.io/language-models/

import random

# starting words
text = ["today", "the"]
sentence_finished = False
 
while not sentence_finished:
  # select a random probability threshold  
  r = random.random()
  accumulator = .0

  for word in model[tuple(text[-2:])].keys():
      accumulator += model[tuple(text[-2:])][word]
      # select words that are above the probability threshold
      if accumulator >= r:
          text.append(word)
          break

  if text[-2:] == [None, None]:
      sentence_finished = True
 
print (' '.join([t for t in text if t]))
```

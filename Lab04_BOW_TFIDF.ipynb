## Feature Engineering: Bag of Words (BOW)
### Using CountVectorizer from sklearn library

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
corpus = ['the car is driven on the road','the truck is driven on the highway']
```

```python
vectorizer = CountVectorizer(min_df=1, ngram_range=(1,1))
features = vectorizer.fit_transform(corpus)

features = features.toarray()
vocab = vectorizer.get_feature_names()
bow = pd.DataFrame(features, columns=vocab)
```

```python
bow
```

# Feature Engineering: Bag of N-grams
### Using CountVectorizer from sklearn library

```python
from sklearn.feature_extraction.text import CountVectorizer
#get bag of words features in sparse format
cv_matrix = CountVectorizer(min_df=0., max_df=1.)
cv_matrix
print(cv_matrix)
```

```python
# you can set the n-gram range to 1,2 to get unigrams as well as bigrams
bv = CountVectorizer(ngram_range=(2,2)) # represent the Bag of bi-gram
bv_matrix = bv.fit_transform(corpus)

bv_matrix = bv_matrix.toarray()
vocab = bv.get_feature_names()
bow_bigram = pd.DataFrame(bv_matrix, columns=vocab)
```

```python
bow_bigram
```

```python
# you can set the n-gram range to 1,2 to get unigrams as well as bigrams
bv = CountVectorizer(ngram_range=(3,3)) # represent the Bag of tri-gram
bv_matrix = bv.fit_transform(corpus)

bv_matrix = bv_matrix.toarray()
vocab = bv.get_feature_names()
bow_trigram = pd.DataFrame(bv_matrix, columns=vocab)
```

# Feature Engineering: TFIDF 
### Using TfidfVectorizer from sklearn library (https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html)

```python
corpus = ['the car is driven on the road','the truck is driven on the highway']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(corpus)
feature_names = vectorizer.get_feature_names()
dense = vectors.todense()
denselist = dense.tolist()
df = pd.DataFrame(denselist, columns=feature_names)
```

```python
df
```

### Using TfidfTransformer from sklearn library (https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfTransformer.html)

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
vectorizer = CountVectorizer()
features = vectorizer.fit_transform(corpus)
tf = pd.DataFrame(features.toarray(), columns=vectorizer.get_feature_names())
```

```python
tf
```

```python
tfidf_transformer = TfidfTransformer()
X = tfidf_transformer.fit_transform(features)
idf = pd.DataFrame({'feature_name':vectorizer.get_feature_names(), 'idf_weights':tfidf_transformer.idf_})
```

```python
idf
```

```python
tf_idf = pd.DataFrame(X.toarray() ,columns=vectorizer.get_feature_names())
```

```python
tf_idf
```
